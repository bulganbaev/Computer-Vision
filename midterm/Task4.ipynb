```python
import matplotlib.pyplot as plt
import matplotlib.image as img

from scipy import ndimage as ndi

from skimage import feature
import cv2
from scipy import misc

from PIL import Image
import numpy as np
%matplotlib inline

image = cv2.imread('./pic.png')
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


row,col= img.shape
gauss = np.random.normal(0,15,(row,col))
gauss = gauss.reshape(row,col)
noisy = img + gauss
img = noisy

plt.imshow(img, cmap ="gray")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
import matplotlib.image as img

from skimage import feature

# Generate noisy image of a square

image = cv2.imread('pic.png',0)
th2 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,7,0)
# Compute the Canny filter for two values of sigma
edges1 = feature.canny(image)
edges2 = feature.canny(image, sigma=2)

# display results
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3),
                                    sharex=True, sharey=True)

ax1.imshow(image, cmap=plt.cm.gray)
ax1.axis('off')
ax1.set_title('noisy image', fontsize=20)

ax2.imshow(edges1, cmap=plt.cm.gray)
ax2.axis('off')
ax2.set_title('Canny filter, $\sigma=1$', fontsize=20)

ax3.imshow(edges2, cmap=plt.cm.gray)
ax3.axis('off')
ax3.set_title('Canny filter, $\sigma=2$', fontsize=20)

fig.tight_layout()

plt.show()
```

```python
#Sobel edge detection
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
from skimage.filters import roberts, sobel, scharr, prewitt


imgNew = img.imread('./pic.png')
#imgNew = color.rgb2gray(image)

edge_sobel = sobel(imgNew[:,:,0])
fig, ax = plt.subplots(ncols=2, sharex=True, sharey=True,
                       figsize=(20, 10))
ax[0].imshow(imgNew, cmap=plt.cm.gray)
ax[0].set_title('Original Image')

ax[1].imshow(edge_sobel, cmap=plt.cm.gray)
ax[1].set_title('Sobel Edge Detection')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```
