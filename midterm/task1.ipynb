```python
import matplotlib.pyplot as plt
import matplotlib.image as img
from scipy import ndimage
import cv2
from scipy import misc

from PIL import Image
import numpy as np
%matplotlib inline

image = cv2.imread('./pic.png')
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


row,col= img.shape
gauss = np.random.normal(0,15,(row,col))
gauss = gauss.reshape(row,col)
noisy = img + gauss
img = noisy

plt.imshow(img, cmap ="gray")
plt.show()
```

```python
##poorly removes noise such as "salt and pepper"
#increases the sharpness of the image 
#A
fig = plt.figure(figsize=(15,10))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3)

k = np.array([[1/16,1/16,1/16],
              [1/16,1/16,1/16],
              [1/16,1/16,1/16]
                 ])
flt = ndimage.convolve(img, k, mode='nearest', cval=0.0)
fmin = flt.min()
fmax = flt.max()
flt = (flt-fmin)/(fmax-fmin)




ax1.set_title('Filter')
ax1.imshow(flt,cmap='gray')

ax2.set_title('Original')
ax2.imshow(img,cmap='gray')

ax3.set_title('Difference')
ax3.imshow(img-flt,cmap='gray')
plt.show()

```

```python
# increases the sharpness of the image
#corresponds to a low-pass filter in the frequency domain;
#Б
fig = plt.figure(figsize=(15,10))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3)
k = np.array([[0,1/5,0],
              [1/5,1/5,1/5],
              [0,1/5,0]
                 ])
flt = ndimage.convolve(img, k, mode='nearest', cval=0.0)
fmin = flt.min()
fmax = flt.max()
flt = (flt-fmin)/(fmax-fmin)




ax1.set_title('Filter')
ax1.imshow(flt,cmap='gray')

ax2.set_title('Original')
ax2.imshow(img,cmap='gray')

ax3.set_title('Difference')
ax3.imshow(img-flt,cmap='gray')
plt.show()

```

```python
#В
#less than the other specified noise-canceling filters, blurs edges; 
#
fig = plt.figure(figsize=(15,10))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3)

flt = ndimage.median_filter(img, 3)
fmin = flt.min()
fmax = flt.max()
flt = (flt-fmin)/(fmax-fmin)




ax1.set_title('Filter')
ax1.imshow(flt,cmap='gray')

ax2.set_title('Original')
ax2.imshow(img,cmap='gray')

ax3.set_title('Difference')
ax3.imshow(img-flt,cmap='gray')
plt.show()
```

```python
#is not a noise canceling filter; 
#Г
#works slower than the other specified noise-canceling filters;
fig = plt.figure(figsize=(15,10))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3)

k = np.array([[-1,-2,-1],
              [-2,12,-2],
              [-1,-2,-1]])
flt = ndimage.convolve(img, k, mode='nearest', cval=0.0)





ax1.set_title('Filter')
ax1.imshow(flt,cmap='gray')

ax2.set_title('Original')
ax2.imshow(img,cmap='gray')

ax3.set_title('Difference')
ax3.imshow(img-flt,cmap='gray')
plt.show()
```
