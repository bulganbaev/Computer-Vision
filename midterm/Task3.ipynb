```python
import matplotlib.pyplot as plt
import matplotlib.image as img
from scipy import ndimage
import cv2
from scipy import misc

from PIL import Image
import numpy as np
%matplotlib inline

image = cv2.imread('./pic.png')
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


row,col= img.shape
gauss = np.random.normal(0,15,(row,col))
gauss = gauss.reshape(row,col)
noisy = img + gauss
fmin = noisy.min()
fmax = noisy.max()
noisy = (noisy-fmin)/(fmax-fmin)


fig = plt.figure(figsize=(15,10))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3)





ax1.set_title('Filter')
ax1.imshow(noisy,cmap='gray')

ax2.set_title('Original')
ax2.imshow(img,cmap='gray')

ax3.set_title('Difference')
ax3.imshow(img-noisy,cmap='gray')
plt.show()
```

```python

fig = plt.figure(figsize=(15,10))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3)

flt = ndimage.median_filter(noisy, 3)
fmin = flt.min()
fmax = flt.max()
flt = (flt-fmin)/(fmax-fmin)




ax1.set_title('Filter')
ax1.imshow(flt,cmap='gray')

ax2.set_title('Original')
ax2.imshow(noisy,cmap='gray')

ax3.set_title('Difference')
ax3.imshow(img-flt,cmap='gray')
plt.show()
```
